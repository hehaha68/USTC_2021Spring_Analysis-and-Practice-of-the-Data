## 总结
根据以上的分析，我们发现：

 1.不论是第一滴血还是第一座防御塔、第一条龙、水晶和先锋，率先获得的队伍的胜率都比未能获得的要高

 2.更进一步的，在影响队伍获胜的概率方面，不同种类夺得先机重要性排序为：
 
  水晶>男爵>防御塔>大龙>峡谷先锋>第一滴血

 3.经济领先是衡量能否取得胜利的重要指标，经济被压制的队伍很难翻盘，更高的经济意味着更容易获胜
  
 4.KDA作为一种后验信息，是对局势的一种量化。在游戏结束之前，或许可以通过当前的KDA数据对最终结果进行估计。从结果上看，获得胜利的队伍的KDA
 
 一般都会很高，反之则较低
 
 5.根据MOBA类游戏的常识，游戏里一般都存在所谓“版本之子”。每次游戏版本更迭，都会出现一部分英雄成为强势角色。仅考虑胜利的前提下，玩家应当
 
 尽量选择顺应潮流的角色。
 
 6.对于热门英雄的胜率未进入前十，我的看法是，热门英雄由于玩的玩家较多，水平参差不齐，在样本较多的情况下胜率逐渐趋于50%。而对于有些冷门的
 
 英雄，一般来说玩他们的都是“绝活”玩家，加之冷门角色很多玩家不熟悉技能，反而更容易取得胜利。一般来说除了拿手的角色外，尽量还是选用热门角
 
 色较稳妥
